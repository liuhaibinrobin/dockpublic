## `NCI_ODDT`

### NCI_ODDT:Noise

```python
nci_oddt_noise = d_noise.groupby("NCI_ODDT").count()[["NCI_ID"]]

nci_oddt_noise.columns = ["Noise_COUNT"]
nci_oddt_noise["Noise_Percentage"] = round(100*nci_oddt_noise["Noise_COUNT"]/sum(nci_oddt_noise["Noise_COUNT"]),5)
nci_oddt_noise.sort_values("NCI_ODDT", ascending=True, inplace = True)
nci_oddt_noise
```

### NCI_ODDT:High or Medium

```python
nci_oddt_hom = d_hom.groupby("NCI_ODDT").count()[["NCI_ID"]]

nci_oddt_hom.columns = ["HoM_COUNT"]
nci_oddt_hom["HoM_Percentage"] = round(100*nci_oddt_hom["HoM_COUNT"]/sum(nci_oddt_hom["HoM_COUNT"]),5)
nci_oddt_hom.sort_values("NCI_ODDT", ascending=True, inplace = True)
nci_oddt_hom
```

### NCI_ODDT:All

```python
nci_oddt_all = d.groupby("NCI_ODDT").count()[["NCI_ID"]]
nci_oddt_all.columns = ["All_COUNT"]
nci_oddt_all["All_Percentage"] = round(100*nci_oddt_all["All_COUNT"]/sum(nci_oddt_all["All_COUNT"]),5)
nci_oddt_all.sort_values("NCI_ODDT", ascending=True, inplace = True)
nci_oddt_all
```

```python
nci_oddt_all_protein = d[d.Group == "protein"].groupby("NCI_ODDT").count()[["NCI_ID"]]
nci_oddt_all_protein.columns = ["All_COUNT_Protein"]
nci_oddt_all_protein["All_Percentage_Protein"] = round(100*nci_oddt_all_protein["All_COUNT_Protein"]/sum(nci_oddt_all_protein["All_COUNT_Protein"]),5)
nci_oddt_all_protein.sort_values("NCI_ODDT", ascending=True, inplace = True)
nci_oddt_all_protein
```
